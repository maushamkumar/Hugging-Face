# Build the text-to-speech pipeline using the 🤗 Transformers Library

- Here is some code that suppresses warning messages.

```python
from transformers.utils import logging

logging.set_verbosity_error()
```

```python
from transformers import pipeline

narrator = pipeline("text-to-speech",
                    model="kakao-enterprise/vits-ljs")
```

```python
text = """Humans are highly intelligent, social beings who belong to the species Homo sapiens.
They are distinguished by their ability to think critically, communicate through complex language,
create tools, and adapt to diverse environments.
Humans have developed cultures, societies, and technologies that shape their way of life and interaction with the world."""
```

```python
narrated_text = narrator(text)
```

```python
from IPython.display import Audio as IPythonAudio

IPythonAudio(narrated_text["audio"][0],
             rate=narrated_text["sampling_rate"])
```
